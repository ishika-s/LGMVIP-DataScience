```python
#importing library
from google.colab.patches import cv2_imshow
import cv2

#reading the image
img=cv2.imread('/content/dp.jpg')

#adjusting the siize opf the image
scale=20
width=int(img.shape[1] * scale/100)
height=int(img.shape[0] * scale/100)
dim=(width,height)
img2=cv2.resize(img,dim,interpolation=cv2.INTER_AREA)
cv2_imshow(img2)
```

```python
#converting into grey scale image
grayimg=cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
cv2_imshow(grayimg)
```

```python
#inverting the image
img_invert = 255-grayimg
cv2_imshow(img_invert)
```

```python
##blurring the image
blurred_img = cv2.GaussianBlur(img_invert, (21, 21), 0)
cv2_imshow(blurred_img)
```

```python
#inverting the blurred image
blurred_invert = 255-blurred_img
cv2_imshow(blurred_invert)
```

```python
#the required pencil sketch
sketch = cv2.divide(grayimg, blurred_invert, scale=240.0)
cv2_imshow(sketch)
```
